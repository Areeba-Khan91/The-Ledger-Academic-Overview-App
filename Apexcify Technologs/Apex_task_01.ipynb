## **Task 1: Rock-Paper-Scissors with AI Opponent**

This task was implemented using a modular Python solution to create a console-based game of Rock-Paper-Scissors. The core objectives—control flow, randomization, and basic game logic—were successfully met.

Implementation Summary:
Core Logic: The game uses a while True loop to manage continuous play and conditional logic (if/elif/else) to determine the winner based on the rules.

Randomization: The random.choice() function from the built-in random module was used to ensure the AI's choice (Rock, Paper, or Scissors) was completely random, meeting the core AI requirement.

User Interaction: A simple input loop handles user choices and validates the input to ensure it is one of the three valid options or the 'quit' command.

Bonus (Advanced AI): An advanced version was developed to demonstrate pattern recognition. This AI tracks the user's entire move history, uses the collections.Counter to find the user's most frequent move (basic probability), and then plays the counter-move to beat the predicted user pattern. This showcased an understanding of basic machine learning logic applied to game design.

```python
import random

def get_player_choice():
    """Prompts the user for their choice and validates the input."""
    choices = ["rock", "paper", "scissors"]
    while True:
        # Get user input and convert to lowercase for case-insensitivity
        user_input = input("Enter your choice (Rock, Paper, Scissors) or 'q' to quit: ").lower()
        if user_input == 'q':
            return 'quit'
        if user_input in choices:
            return user_input
        else:
            print("Invalid choice. Please enter 'Rock', 'Paper', or 'Scissors'.")

def get_computer_choice_random(choices):
    """Computer makes a completely random choice using the random module."""
    return random.choice(choices)

def determine_winner(player_move, computer_move):
    """Applies the game logic (if/else/elif) to determine the winner."""
    if player_move == computer_move:
        return "Tie"

    # Win conditions
    elif (player_move == "rock" and computer_move == "scissors") or \
         (player_move == "paper" and computer_move == "rock") or \
         (player_move == "scissors" and computer_move == "paper"):
        return "Player"

    # All other cases are losses
    else:
        return "Computer"

def rock_paper_scissors_random_game():
    """Main function for the basic game loop."""
    choices = ["rock", "paper", "scissors"]
    player_score = 0
    computer_score = 0

    print("\n--- Rock-Paper-Scissors: Random AI Opponent ---")

    # Game Loop
    while True:
        player_move = get_player_choice()

        if player_move == 'quit':
            break

        computer_move = get_computer_choice_random(choices)

        # Display moves
        print(f"\nPlayer chose: {player_move.capitalize()}")
        print(f"Computer chose: {computer_move.capitalize()}")

        # Determine and display result
        result = determine_winner(player_move, computer_move)

        if result == "Player":
            player_score += 1
            print("You Win! 🎉")
        elif result == "Computer":
            computer_score += 1
            print("You Lose! 🤖")
        else:
            print("It's a Tie! 🤝")

        print(f"Score: Player {player_score} - Computer {computer_score}\n")

    print(f"\n--- Final Score: Player {player_score} - Computer {computer_score} ---")
    print("Thanks for playing!")

# To run the core game:
rock_paper_scissors_random_game()
```

```python
from collections import Counter
import random

# Reuse the determine_winner and get_player_choice functions from the core version
# ... (for brevity, assume they are included or copied above) ...

def get_computer_choice_learning(choices, player_history):
    """
    Computer AI learns user patterns using basic probability.

    1. Tracks user's history.
    2. Finds the user's most frequent move (prediction).
    3. Chooses the counter-move.
    """
    if len(player_history) < 3: # Need some data, start with a random move for the first few rounds
        return random.choice(choices)

    # 1. Tally the moves
    move_counts = Counter(player_history)

    # 2. Find the most common move (the AI's prediction of what the player will do next)
    # The [0][0] gets the move name from the list of tuples returned by most_common()
    most_frequent_move = move_counts.most_common(1)[0][0]

    print(f"🤖 AI analyzed history. Player's most frequent move: {most_frequent_move.capitalize()}")

    # 3. Determine the counter-move to beat the prediction
    counter_moves = {
        "rock": "paper",    # Paper beats Rock
        "paper": "scissors", # Scissors beats Paper
        "scissors": "rock"   # Rock beats Scissors
    }

    ai_choice = counter_moves[most_frequent_move]
    return ai_choice

def rock_paper_scissors_learning_game():
    """Main function for the Advanced Learning AI game loop."""
    choices = ["rock", "paper", "scissors"]
    player_score = 0
    computer_score = 0
    player_history = [] # This list stores every move the player makes

    print("\n--- Rock-Paper-Scissors: Learning AI Opponent (Bonus) ---")
    print("The AI will try to beat your most common move. Play 3+ rounds for it to start learning!")

    # Game Loop
    while True:
        player_move = get_player_choice()

        if player_move == 'quit':
            break

        # AI chooses its move based on history (or randomly if history is short)
        computer_move = get_computer_choice_learning(choices, player_history)

        # Add the player's move to history *after* the AI has made its choice for the current round
        player_history.append(player_move)

        # Display moves
        print(f"\nPlayer chose: {player_move.capitalize()}")
        print(f"Computer chose: {computer_move.capitalize()}")

        # Determine and display result
        result = determine_winner(player_move, computer_move)

        if result == "Player":
            player_score += 1
            print("You Win! 🎉")
        elif result == "Computer":
            computer_score += 1
            print("You Lose! 🤖 (The AI's prediction was correct!)")
        else:
            print("It's a Tie! 🤝")

        print(f"Score: Player {player_score} - Computer {computer_score}\n")

    print(f"\n--- Final Score: Player {player_score} - Computer {computer_score} ---")
    print("Thanks for playing!")

# To run the learning game, uncomment the line below:
rock_paper_scissors_learning_game()
```
